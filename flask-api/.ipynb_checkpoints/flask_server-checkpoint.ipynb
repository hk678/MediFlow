````python
from flask import Flask, request, jsonify
from dotenv import load_dotenv
from langchain.prompts import PromptTemplate
from langchain_openai import ChatOpenAI
import os
import time
import json
import re
from langchain.schema import AIMessage
from collections import OrderedDict  # íŒŒì¼ ìƒë‹¨ import!

load_dotenv()
api_key = os.getenv("OPENAI_API_KEY")


app = Flask(__name__)

# LangChainì—ì„œ PromptTemplateì„ ì¨ì„œ, í”„ë¡¬í”„íŠ¸ + ë³€ìˆ˜ ê°’ ë„£ëŠ” í˜•ì‹ìœ¼ë¡œ ë°”ê¿ˆ
prompt_template = PromptTemplate(
    input_variables=[
        "gender", "age", "acuity", "pain", "chief_complaint", "arrival_transport",
        "HR", "RR", "SpO2", "SBP", "DBP", "BT",
        "hemoglobin", "wbc", "plateletCount", "redBloodCells", "sedimentationRate",
        "na", "k", "chloride", "ca", "mg", "ureaNitrogen", "creatinine",
        "ast", "alt", "bilirubin", "albumin", "ap", "ggt", "ld", "ammonia",
        "glucose", "lactate", "acetone", "bhb", "crp", "pt", "inrPt", "ptt",
        "dDimer", "troponinT", "ck", "ckmb", "ntprobnp", "amylase", "lipase",
        "ph", "pco2", "po2", "ctco2", "bcb"
    ],
    template="""
ë„ˆëŠ” ê²½ë ¥ 15ë…„ ì´ìƒì˜ ì‘ê¸‰ì˜í•™ê³¼ ì „ë¬¸ì˜ì´ë‹¤.
ì•„ë˜ í™˜ì ì •ë³´ì—ì„œ 'acuity'ëŠ” ë¯¸êµ­ ESI(1~5) ë“±ê¸‰ì´ë‹ˆ, ì´ë¥¼ ì°¸ê³ í•´ ì˜ˆìƒë˜ëŠ” KTAS ë“±ê¸‰(1~5)ë¡œ ì„ì‹œ ë§¤í•‘í•˜ì—¬ ì ìˆ˜ë¥¼ ì‚°ì •í•˜ë¼.
(ì°¸ê³ : ESIì™€ KTASëŠ” í™˜ì ë¶„ë¥˜ ì„¸ë¶€ê¸°ì¤€ì€ ë‹¤ë¥´ì§€ë§Œ, ìˆ«ì ë°©í–¥ì„±(1=ì¤‘ì¦, 5=ê²½ì¦)ì€ ë™ì¼í•¨)

[ì„ì‹œ ë§¤í•‘í‘œ]
- acuity 1 â†’ KTAS 1ë“±ê¸‰
- acuity 2 â†’ KTAS 2ë“±ê¸‰
- acuity 3 â†’ KTAS 3ë“±ê¸‰
- acuity 4 â†’ KTAS 4ë“±ê¸‰
- acuity 5 â†’ KTAS 5ë“±ê¸‰

ë‹¤ìŒì˜ ì˜í•™ì  ê¸°ì¤€ì„ ë°˜ë“œì‹œ ì§€ì¼œì„œ ì ìˆ˜ë¥¼ ë§¤ê¸¸ ê²ƒ:

â‘  ë‚˜ì´: 65ì„¸ ì´ìƒ(+10ì , ê³ ë ¹ í™˜ì ìƒíƒœ ì•…í™” ê°€ëŠ¥ì„± ë†’ìŒ)  
â‘¡ ì£¼ìš” ì¦ìƒ: í‰í†µ, ì˜ì‹ì €í•˜, í˜¸í¡ê³¤ë€(+20ì , ì¤‘ì¦ ì§ˆí™˜ ì‹ í˜¸)  
â‘¢ í™œë ¥ì§•í›„ ì´ìƒ: 
  - í˜ˆì••ì €í•˜(SBP<90, DBP<60), 
  - ë¹ˆë§¥(HR>100), ì„œë§¥(HR<60),
  - í˜¸í¡ê³¤ë€(RR>24 ë˜ëŠ” <12), 
  - SpO2<94%, 
  - ì²´ì˜¨ ì´ìƒ(BT>38ë„ ë˜ëŠ” <36ë„)
  ê° í•­ëª© ë°œê²¬ ì‹œ +15ì , ë³µìˆ˜ í•´ë‹¹ ì‹œ ì¤‘ë³µê°€ì‚°
â‘£ í†µì¦(NRS ê¸°ì¤€): ì¤‘ì¦(8~10ì )+10ì , ì¤‘ë“±ë„(4~7ì )+5ì   
â‘¤ ë„ì°© ê²½ë¡œ: 119 êµ¬ê¸‰ì°¨ ì´ì†¡(+5ì , ì‘ê¸‰ í™˜ìëŠ” ì¤‘ì¦ ê°€ëŠ¥ì„± ë†’ìŒ)  
â‘¥ KTAS ë“±ê¸‰(=ESI ì„ì‹œ ë§¤í•‘): 1ë“±ê¸‰(+30ì ), 2ë“±ê¸‰(+20ì ), 3ë“±ê¸‰(+10ì ), 4~5ë“±ê¸‰(0ì )

ì ìˆ˜ ê¸°ì¤€ disposition (ë°˜ë“œì‹œ ì´ ì¤‘ í•˜ë‚˜ë§Œ ì„ íƒ, ë‹¤ë¥¸ í‘œí˜„ ì ˆëŒ€ ê¸ˆì§€!)
- 1~33ì : "ê·€ê°€"
- 34~66ì : "ì¼ë°˜ë³‘ë™" (ì ˆëŒ€ ê³¼ëª…, ê´€ì°°ì‹¤ ë“±ì˜ í‘œí˜„ ì‚¬ìš© ê¸ˆì§€!)
- 67~100ì : "ICU"

ë°˜ë“œì‹œ ë‹¤ìŒ JSON í˜•ì‹ìœ¼ë¡œë§Œ ë‹µë³€ (ìˆœì„œ ë° í‚¤ ë³€ê²½ ê¸ˆì§€):
{{
  "risk_score": (1~100 ì •ìˆ˜),
  "disposition": (0, 1, 2 ì¤‘ í•˜ë‚˜ë§Œ, 0=ê·€ê°€, 1=ì¼ë°˜ë³‘ë™, 2=ICU),
  "clinical_reason": "(ê° ìš”ì†Œì˜ ì ìˆ˜ ê·¼ê±°ë¥¼ í¬í•¨í•´, ESIâ†’KTAS ì„ì‹œ ë§¤í•‘ì„ì„ ì„¤ëª…í•˜ë©° 1~2ë¬¸ì¥)"
}}

clinical_reason ì˜ˆì‹œ:
"acuity(ESI 2â†’KTAS 2ë“±ê¸‰, +20), í‰í†µ(+20), ê³ ë ¹(+10) ë“± ì¤‘ì¦ ìš”ì†Œ ë°˜ì˜. ESIì™€ KTASì˜ ê¸°ì¤€ ì°¨ì´ë¡œ ì‹¤ì œ ë“±ê¸‰ì€ ë‹¤ë¥¼ ìˆ˜ ìˆìŒ."

ì•„ë˜ëŠ” ì˜ëª»ëœ ì˜ˆì‹œ (ì ˆëŒ€ ì´ë ‡ê²Œ ë‹µí•˜ì§€ ë§ˆì‹œì˜¤):
{{
  "risk_score": 45,
  "disposition": "ì •ì‹ ê³¼ ë³‘ë™", â† ì˜¤ë‹µ!
  "clinical_reason": "ê³ ë ¹ìœ¼ë¡œ ì…ì› í•„ìš”ì„± ë†’ìŒ."
}}

ì˜¬ë°”ë¥¸ ì˜ˆì‹œ:
{{
  "risk_score": 45,
  "disposition": 1,
  "clinical_reason": "65ì„¸ ê³ ë ¹(+10), ì¤‘ë“±ë„ í†µì¦(+5), KTAS 3ë“±ê¸‰(+10), ì…ì› í•„ìš”ì„± ìˆìŒ. ê³ ë ¹ í™˜ìëŠ” ìƒíƒœ ê¸‰ë³€ ê°€ëŠ¥, ì¤‘ë“±ë„ ì¦ìƒì€ ì¶”ê°€ í‰ê°€ í•„ìš”."
}}



sample_outputs = [
    {{
        "input": {{
            "gender": 0,
            "age": 55,
            "acuity": None,
            "pain": None,
            "chief_complaint": "LEFT HAND PAINS",
            "arrival_transport": "WALK IN"
        }},
        "output": {{
            "risk_level": "ì¤‘ê°„",
            "disposition": 2,  
            "clinical_reason": "55ì„¸ ë‚¨ì„±ìœ¼ë¡œ, left hand painì„ ì£¼ì†Œë¡œ ë‚´ì›. vital ì•ˆì •ì ì´ë‚˜ ì§„ë‹¨ ë° ì¤‘ì¦ë„ í‰ê°€ìƒ ì¤‘í™˜ìì‹¤ ì§‘ì¤‘ê´€ì°° í•„ìš”."
        }}
    }},
    {{
        "input": {{
            "gender": 0,
            "age": 55,
            "acuity": 3,
            "pain": None,
            "chief_complaint": "LEFT HAND PAINS",
            "arrival_transport": "WALK IN"
        }},
        "output": {{
            "risk_level": "ì¤‘ê°„",
            "disposition": 2,
            "clinical_reason": "left hand pain ë° fever ì¦ìƒ. ì¦ìƒ ë° ê´€ì°° í•„ìš”ì„±ìœ¼ë¡œ ICU ë°°ì •."
        }}
    }},
    {{
        "input": {{
            "gender": 0,
            "age": 55,
            "acuity": 3,
            "pain": None,
            "chief_complaint": "LEFT HAND PAINS",
            "arrival_transport": "WALK IN"
        }},
        "output": {{
            "risk_level": "ì‹¬ê°",
            "disposition": 2,
            "clinical_reason": "ê´€ì°°ì‹¤ì—ì„œ ëª¨ë‹ˆí„°ë§ í›„ ì¤‘ì¦ ì§ˆí™˜ ê°€ëŠ¥ì„± ìˆì–´ ì¤‘í™˜ìì‹¤ ì§‘ì¤‘ê´€ì°° í•„ìš”."
        }}
    }},
    {{
        "input": {{
            "gender": 0,
            "age": 55,
            "acuity": 3,
            "pain": None,
            "chief_complaint": "LEFT HAND PAINS",
            "arrival_transport": "WALK IN"
        }},
        "output": {{
            "risk_level": "ì‹¬ê°",
            "disposition": 2,
            "clinical_reason": "í™˜ì ìƒíƒœ ê³ ë ¤ ì‹œ ê¸‰ì„± ì•…í™” ê°€ëŠ¥ì„±ì´ ìˆì–´ ICUì—ì„œ ì¶”ê°€ í‰ê°€ ê¶Œì¥."
        }}
    }},
    {{
        "input": {{
            "gender": 0,
            "age": 55,
            "acuity": 3,
            "pain": 3,
            "chief_complaint": "LEFT HAND PAINS",
            "arrival_transport": "WALK IN"
        }},
        "output": {{
            "risk_level": "ì¤‘ê°„",
            "disposition": 2,
            "clinical_reason": "ì¤‘ë“±ë„ ì† í†µì¦ ë° ë™ë°˜ ì¦ìƒ ì§€ì†, ì¤‘í™˜ìì‹¤ì—ì„œ ëª¨ë‹ˆí„°ë§ í•„ìš”."
        }}
    }}
]



{{
  "risk_level": "ë‚®ìŒ",
  "disposition": 1,
  "clinical_reason": "ì—¬ì„±, ë„ë³´ë¡œ ë‚´ì›, ê²½ë¯¸í•œ í˜¸í¡ê³¤ë€ ì™¸ì— íŠ¹ì´ ì¦ìƒì´ë‚˜ ìœ„í—˜ ìš”ì¸ ì—†ì–´ ê·€ê°€ ì¡°ì¹˜ ê°€ëŠ¥"
}}

ì•„ë˜ í™˜ì ì •ë³´ë¥¼ ì°¸ê³ í•´ ë°˜ë“œì‹œ ë™ì¼ í˜•ì‹ìœ¼ë¡œ ë‹µë³€í•´.  
ì¡°ê±´ì„ ë‹¨ í•˜ë‚˜ë¼ë„ ì–´ê¸°ë©´ 0ì ì´ë‹ˆ ì£¼ì˜í•´.

í™˜ì ì •ë³´:
- ì„±ë³„: {gender}
- ë‚˜ì´: {age}
- acuity(ESI): {acuity}
- í†µì¦(NRS): {pain}
- ì£¼ìš” ì¦ìƒ: {chief_complaint}
- ë„ì°© ê²½ë¡œ: {arrival_transport}
- ë§¥ë°•(HR): {HR}
- í˜¸í¡ìˆ˜(RR): {RR}
- ì‚°ì†Œí¬í™”ë„(SpO2): {SpO2}
- ìˆ˜ì¶•ê¸°í˜ˆì••(SBP): {SBP}
- ì´ì™„ê¸°í˜ˆì••(DBP): {DBP}
- ì²´ì˜¨(BT): {BT}
"""
)

# LangChainì€ LLM ê°ì²´ë¡œ OpenAI ì—°ë™
llm = ChatOpenAI(
    openai_api_key=os.getenv("OPENAI_API_KEY"),
    model="gpt-4o",
    temperature=0.0   # â† ì—¬ê¸°ì—ì„œ 0.0~0.2 ì •ë„ë¡œ ì£¼ë©´ ë¨!
)

#chain êµ¬ì„±
# íŒŒì´í”„(|) ì—°ì‚°ì ì‚¬ìš©!
chain = prompt_template | llm


def clean_json_codeblock(s):
    s = s.strip()
    if s.startswith("```"):
        s = re.sub(r"^```json\s*|```$", "", s, flags=re.MULTILINE).strip()
    return s

def validate_and_adjust_result(result_dict, input_data):
    """ê²°ê³¼ ê²€ì¦ ë° ì¡°ì •"""
    
    # ì ìˆ˜ ë²”ìœ„ í™•ì¸
    score = result_dict.get("risk_score", 0)
    if score < 0:
        score = 0
    elif score > 100:
        score = 100
    
    # Disposition ì¬ê²€ì¦
    disposition = result_dict.get("disposition")
    
    # ëª…ë°±í•œ ì˜¤íŒ êµì •
    chief_complaint = str(input_data.get('chief_complaint', '')).upper()
    age = input_data.get('age', 0)
    acuity = input_data.get('acuity')
    
    # ê²½ë¯¸í•œ ì¦ìƒë“¤
    minor_symptoms = ['HAND PAIN', 'FINGER PAIN', 'MINOR LACERATION', 'COMMON COLD', 
                     'HEADACHE', 'MINOR ABRASION', 'ANKLE SPRAIN']
    
    is_minor = any(symptom in chief_complaint for symptom in minor_symptoms)
    
    # í™œë ¥ì§•í›„ í™•ì¸
    hr = input_data.get('HR')
    sbp = input_data.get('SBP') 
    spo2 = input_data.get('SpO2')
    
    vital_stable = True
    if hr and (hr < 50 or hr > 120):
        vital_stable = False
    if sbp and (sbp < 90 or sbp > 180):
        vital_stable = False
    if spo2 and spo2 < 90:
        vital_stable = False
    
    # ì˜¤íŒ êµì • ë¡œì§
    if is_minor and vital_stable and age < 65 and acuity and acuity >= 4:
        if disposition == 2:  # ICUë¡œ ì˜ëª» íŒì •ëœ ê²½ìš°
            disposition = 0  # ê·€ê°€ë¡œ êµì •
            score = min(score, 35)
    
    # ìœ„í—˜ ì‹ í˜¸ê°€ ìˆëŠ”ë° ê·€ê°€ë¡œ íŒì •ëœ ê²½ìš°
    if (not vital_stable or age >= 80 or (acuity and acuity <= 2)) and disposition == 0:
        disposition = max(disposition, 1)  # ìµœì†Œ ì¼ë°˜ë³‘ë™
    
    # ì ìˆ˜ì™€ disposition ì¼ì¹˜ì„± í™•ì¸
    if score >= 85 and disposition < 2:
        disposition = 2
    elif score >= 45 and disposition == 0:
        disposition = 1
    elif score < 45 and disposition > 1:
        disposition = min(1, disposition)
    
    return {
        "risk_score": int(score),
        "disposition": int(disposition),
        "clinical_reason": result_dict.get("clinical_reason", "")
    }

@app.route('/predict/admission', methods=['POST'])
def predict():
    try:
        data = request.get_json() # í”„ë¡ íŠ¸ì—ì„œ ë³´ë‚¸ JSON ë°ì´í„° ë°›ê¸°
        result = chain.invoke({
            "gender": data['gender'],
            "age": data['age'],
            "acuity": data['acuity'],
            "pain": data['pain'],
            "chief_complaint": data['chief_complaint'],
            "arrival_transport": data['arrival_transport'],
            "HR": data.get('HR', None),
            "RR": data.get('RR', None),
            "SpO2": data.get('SpO2', None),
            "SBP": data.get('SBP', None),
            "DBP": data.get('DBP', None),
            "BT": data.get('BT', None),
        })
                
            # AIMessage â†’ content ë¬¸ìì—´ë¡œ ë³€í™˜ (í•µì‹¬!!)
        if isinstance(result, AIMessage):
            result_str = result.content
        else:
            result_str = str(result)

        print("LLM ê²°ê³¼:", result_str)

        result_str = clean_json_codeblock(result_str)
        result_dict = json.loads(result_str)

          # ** dispositionì´ ë¬¸ìì—´ì´ë©´ ì—ëŸ¬! (ì—¬ê¸°ì„œ ë°”ë¡œ ê²€ì¦) **
        if not isinstance(result_dict.get("disposition"), int):
            return jsonify({"error": "dispositionì€ ë°˜ë“œì‹œ ìˆ«ì(0,1,2)ì—¬ì•¼ í•©ë‹ˆë‹¤.", "result": result_dict}), 500

               

            # ì—¬ê¸°ì— ìˆœì„œ ë§ì¶”ëŠ” ë¶€ë¶„ ì¶”ê°€!
        ordered_result = OrderedDict([
            ("risk_score", result_dict.get("risk_score")),
            ("disposition", result_dict.get("disposition")),
            ("clinical_reason", result_dict.get("clinical_reason")),
        ])
        # ğŸ‘‡ JSON ë¬¸ìì—´ë¡œ ì§ì ‘ ë¦¬í„´!
        return jsonify({"result": ordered_result})
                       
    except Exception as e:
        print("json.loads error:", e)
        # ê²°ê³¼ ê²€ì¦ ë° ì¡°ì •
        validated_result = validate_and_adjust_result(result_dict, data)

        # dispositionì´ ì •ìˆ˜ì¸ì§€ í™•ì¸
        if not isinstance(validated_result.get("disposition"), int):
            return jsonify({"error": "dispositionì€ ë°˜ë“œì‹œ ìˆ«ì(0,1,2)ì—¬ì•¼ í•©ë‹ˆë‹¤.", "result": validated_result}), 500

        # ìˆœì„œ ë§ì¶”ê¸° (promptNumì„ ë°±ì—”ë“œì—ì„œ ì§ì ‘ ì¶”ê°€)
        ordered_result = OrderedDict([
            ("risk_score", validated_result.get("risk_score")),
            ("disposition", validated_result.get("disposition")),
            ("clinical_reason", validated_result.get("clinical_reason")),
            ("promptNum", 1),  # í˜„ì¬ í”„ë¡¬í”„íŠ¸ ë²ˆí˜¸ ì§ì ‘ ì§€ì •
        ])
        
        return jsonify({"result": ordered_result})
                       
    except Exception as e:
        print("Error:", e)
        return jsonify({"error": "ì„œë²„ ë‚´ë¶€ ì˜¤ë¥˜", "detail": str(e)}), 500

@app.route('/predict/discharge', methods=['POST'])
def predict_discharge():
    try:
        data = request.get_json()

        # PatientRequestì—ì„œ ì“¸ ìˆ˜ ìˆëŠ” ëª¨ë“  ê°’ì„ ë°›ì•„ì„œ promptì— ë„£ì„ ìˆ˜ ìˆìŒ
        input_vars = {
            "gender": data['gender'],
            "age": data['age'],
            "acuity": data['acuity'],
            "pain": data['pain'],
            "chief_complaint": data['chief_complaint'],
            "arrival_transport": data['arrival_transport'],
            "HR": data.get('HR', None),
            "RR": data.get('RR', None),
            "SpO2": data.get('SpO2', None),
            "SBP": data.get('SBP', None),
            "DBP": data.get('DBP', None),
            "BT": data.get('BT', None),

            # í˜ˆì•¡ ê²€ì‚¬ ê²°ê³¼ í•­ëª©ë“¤ë„ ë‹¤ í¬í•¨í•´ì„œ ì „ë‹¬
            "hemoglobin": data.get('hemoglobin', None),
            "wbc": data.get('wbc', None),
            "plateletCount": data.get('plateletCount', None),
            "redBloodCells": data.get('redBloodCells', None),
            "sedimentationRate": data.get('sedimentationRate', None),
            "na": data.get('na', None),
            "k": data.get('k', None),
            "chloride": data.get('chloride', None),
            "ca": data.get('ca', None),
            "mg": data.get('mg', None),
            "ureaNitrogen": data.get('ureaNitrogen', None),
            "creatinine": data.get('creatinine', None),
            "ast": data.get('ast', None),
            "alt": data.get('alt', None),
            "bilirubin": data.get('bilirubin', None),
            "albumin": data.get('albumin', None),
            "ap": data.get('ap', None),
            "ggt": data.get('ggt', None),
            "ld": data.get('ld', None),
            "ammonia": data.get('ammonia', None),
            "glucose": data.get('glucose', None),
            "lactate": data.get('lactate', None),
            "acetone": data.get('acetone', None),
            "bhb": data.get('bhb', None),
            "crp": data.get('crp', None),
            "pt": data.get('pt', None),
            "inrPt": data.get('inrPt', None),
            "ptt": data.get('ptt', None),
            "dDimer": data.get('dDimer', None),
            "troponinT": data.get('troponinT', None),
            "ck": data.get('ck', None),
            "ckmb": data.get('ckmb', None),
            "ntprobnp": data.get('ntprobnp', None),
            "amylase": data.get('amylase', None),
            "lipase": data.get('lipase', None),
            "ph": data.get('ph', None),
            "pco2": data.get('pco2', None),
            "po2": data.get('po2', None),
            "ctco2": data.get('ctco2', None),
            "bcb": data.get('bcb', None),
        }

        result = chain.invoke(input_vars)
        # ì•„ë˜ëŠ” ë™ì¼!
        input_vars = {
            "gender": data.get('gender', 0),
            "age": data.get('age', 0),
            "acuity": data.get('acuity'),
            "pain": data.get('pain'),
            "chief_complaint": data.get('chief_complaint', ''),
            "arrival_transport": data.get('arrival_transport', ''),
            "HR": data.get('HR'),
            "RR": data.get('RR'),
            "SpO2": data.get('SpO2'),
            "SBP": data.get('SBP'),
            "DBP": data.get('DBP'),
            "BT": data.get('BT'),
            # ëª¨ë“  ê²€ì‚¬ ê²°ê³¼
            "hemoglobin": data.get('hemoglobin'),
            "wbc": data.get('wbc'),
            "plateletCount": data.get('plateletCount'),
            "redBloodCells": data.get('redBloodCells'),
            "sedimentationRate": data.get('sedimentationRate'),
            "na": data.get('na'),
            "k": data.get('k'),
            "chloride": data.get('chloride'),
            "ca": data.get('ca'),
            "mg": data.get('mg'),
            "ureaNitrogen": data.get('ureaNitrogen'),
            "creatinine": data.get('creatinine'),
            "ast": data.get('ast'),
            "alt": data.get('alt'),
            "bilirubin": data.get('bilirubin'),
            "albumin": data.get('albumin'),
            "ap": data.get('ap'),
            "ggt": data.get('ggt'),
            "ld": data.get('ld'),
            "ammonia": data.get('ammonia'),
            "glucose": data.get('glucose'),
            "lactate": data.get('lactate'),
            "acetone": data.get('acetone'),
            "bhb": data.get('bhb'),
            "crp": data.get('crp'),
            "pt": data.get('pt'),
            "inrPt": data.get('inrPt'),
            "ptt": data.get('ptt'),
            "dDimer": data.get('dDimer'),
            "troponinT": data.get('troponinT'),
            "ck": data.get('ck'),
            "ckmb": data.get('ckmb'),
            "ntprobnp": data.get('ntprobnp'),
            "amylase": data.get('amylase'),
            "lipase": data.get('lipase'),
            "ph": data.get('ph'),
            "pco2": data.get('pco2'),
            "po2": data.get('po2'),
            "ctco2": data.get('ctco2'),
            "bcb": data.get('bcb'),
        }

        result = chain.invoke(input_vars)
        
        if isinstance(result, AIMessage):
            result_str = result.content
        else:
            result_str = str(result)
        result_str = clean_json_codeblock(result_str)
        result_dict = json.loads(result_str)
        if not isinstance(result_dict.get("disposition"), int):
            return jsonify({"error": "dispositionì€ ë°˜ë“œì‹œ ìˆ«ì(0,1,2)ì—¬ì•¼ í•©ë‹ˆë‹¤.", "result": result_dict}), 500
        ordered_result = OrderedDict([
            ("risk_score", result_dict.get("risk_score")),
            ("disposition", result_dict.get("disposition")),
            ("clinical_reason", result_dict.get("clinical_reason")),
        ])
        return jsonify({"result": ordered_result})
    except Exception as e:
        print("json.loads error:", e)
        return jsonify({"error": "ì„œë²„ ë‚´ë¶€ ì˜¤ë¥˜", "detail": str(e)}), 500


if __name__ == '__main__':
    app.run(debug=True)
````
